```
a= []
n = int(input())
a = list (map (int,input().split()))
a.sort()
res= 0 
for i in range(n-1):
  if a[i]+1 != a[i+1]: 
    res = res + a[i+1]-a[i]-1
print(res)
```

```
n,m  = map(int, input().split())
a = list (map (int,input().split()))
res = 0
for i in range(n):
  if a[i] >= a[m]:
    res = res+1
print(res)
```
